```python
import pandas as pd
from datasets  import load_dataset
import  matplotlib.pyplot   as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_US = df[df['job_country'] == 'United States'].copy()
```

```python
df_US
```

```python
df_US['job_posted_month'] = df['job_posted_date'].dt.strftime('%B')
```

```python
df_US_pivot = df_US.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
```

```python
df_US_pivot.reset_index(inplace= True)
df_US_pivot['month_no'] = pd.to_datetime(df_US_pivot['job_posted_month'], format= '%B').dt.month
df_US_pivot.sort_values('month_no', inplace= True)
df_US_pivot.set_index('job_posted_month', inplace= True)
df_US_pivot.drop(columns='month_no', inplace= True)
```

```python
df_US_pivot
```

```python

df_US_pivot.plot(kind=('line'))
```

```python
top_3 = df_US['job_title_short'].value_counts().head(3)
top_3 = top_3.index.to_list()
```

```python
df_US_pivot[top_3].plot(kind='line')
plt.title('Monthly Job Postings for Data Jobs in the US')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.legend()
plt.show()

```

```python
df_US_softwares_pivot = pd.read_csv("https://lukeb.co/software_csv", index_col= 'job_posted_month')
df_US_softwares_pivot   
```

```python
df_US_Merge = df_US_pivot.merge(df_US_softwares_pivot, on="job_posted_month")
```

```python
top_5 = (df_US_Merge.
         sum().
         sort_values(ascending=False).
         head().
         index.
         to_list())
top_5

```

```python
df_US_Merge[top_5].plot(kind='line')
plt.title('Monthly Job Postings for Top Tech Jobs in the US')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.ylim(0, 20000)
plt.show()
```
